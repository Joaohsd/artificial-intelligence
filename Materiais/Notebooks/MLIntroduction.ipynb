# **Um pouco sobre os Dados**

Entre os tipos de dados, podemos defini-los como sendo:


*   **Numéricos:** Dados quantitativos;
    *   **Discreto:** Inteiros;
    *   **Contínuos:** Reais;
*   **Categóricos:** Dados que não podem ser comparados entre si (ex.: Cores);
*   **Ordinários:** Dados que poder ser compardos entre si (ex.: Série na escola);





Entre as características dos **dados numéricos**, podemos extrair inúmeras, tais quais:


*   **Média:** Valor médio de um conjunto de dados;
*   **Mediana:** Valor central de um conjunto de dados **ordenados**;
*   **Moda:** Valor mais presente em um conjunto de dados;
*   **Desvio Padrão:** Descreve quanto os valores estão espalhados em relação ao valor médio do conjunto;

    *   **Desvio padrão baixo:** Os valores estão mais próximos da média do conjunto.
    *   **Desvio padrão alto:** Os valores estão mais espalhados em relação a média do conjunto;

*   **Variância:** Outra forma de verificar o quão espalhado os dados estão;









# **Distibuição dos dados**

No mundo real, existem muitos dados. Para isso, é necessário verificar seu conjunto e, muitas vezes, analisar o comportamento desses dados quanto a sua frequência dentro deste conjunto. Dessa forma, a probabilidade é um conceito muito importante no Aprendizado de Máquina.



## **Distribuição Uniforme**

Na **distribuição uniforme**, a probabilidade de se gerar qualquer ponto em um intervalo contido no espaço amostral é proporcional ao tamanho do intervalo, visto que na distribuição uniforme a f(x) é igual para qualquer valor de x no intervalo considerado. Por exemplo, se considerarmos um intervalo em x de zero à dez positivo (xЄ[0,10] ), e assumirmos que temos uma distribuição uniforme nesse intervalo, a probabilidade de f(x) no intervalo  2<x<5 é igual a probabilidade de f(x) no intervalo  5<x<8, pois sabemos que a distribuição é uniforme nesses intervalos e possuímos os intervalos com o mesmo tamanho. A função densidade de probabilidade é definida abaixo:

> image.png



```python
import matplotlib.pyplot as plt

x = numpy.random.uniform(5.0, 1, 1000)
y = numpy.random.uniform(5.0, 1, 1000)

plt.scatter(x, y)
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 100000)

plt.xlabel("Eixo x")
plt.ylabel("Eixo Y")
plt.title("Histograma Uniforme")

plt.hist(x, 100)
plt.show()
```

## **Distribuição Normal (Gaussiana)**

A **distribuição normal** é uma das distribuições de probabilidade mais utilizadas para modelar fenômenos naturais. Isso se deve ao fato de que um grande número de fenômenos naturais apresenta sua distribuição de probabilidade tão proximamente normal, que a ela pode ser com sucesso referida, e, portanto, com adequado acerto por ela representada como se normal fosse. A função densidade de probabilidade é definida abaixo:



> image.png







```python
import matplotlib.pyplot as plt

# Criando duas variáveis aleatórias de 1000 amostras com distribuição gaussiana.  
# Esssa possui média 5 e desvio padrão 1

x = numpy.random.normal(5.0, 1, 1000)
y = numpy.random.normal(5.0, 1, 1000)

plt.scatter(x, y)
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt

# Criando uma variável aleatória de 100000 amostras com distribuição gaussiana.  
# Esssa possui média 5 e desvio padrão 1

x = numpy.random.normal(5.0, 1.0, 100000)

plt.hist(x, 100)
plt.show() 
```

## Aprendizado Supervisionado

O **aprendizado supervisionado** é um paradigma de aprendizado de máquina, que tem como objetivo adquirir informações de relacionamento entre entrada e saída de um sistema, baseado em um conjunto de amostras de treinamento. 

Um algoritmo de aprendizado supervisionado analisa os dados de treino e produz uma função inferida que será utilizada para mapear novos exemplos. Para deixar menos abstrato, a classificação de emails como Spam ou não Spam são utilizam técnicas de aprendizado supervisionado.

### Classificação

A classificação é o processo de categorizar um determinado conjunto de dados em classes. No exemplo da classificação de e-mails como spam, teríamos um exemplo de classificação binária, no qual o modelo através dos dados fornecidos, precisaria gerar como resposta se o e-mail é spam ou não.

Alguns dos algoritmos mais famosos são:

*    KNN
*    Naive Bayes
*    Logistic Regression
*    Support Vector Machines
*    Decision Trees

### Regressão

Os modelos de regressão são utilizados quando queremos prever valores, por exemplo, prever o preço de uma casa ou o número de produtos que serão vendidos em determinado mês.
Alguns modelos são famosos para realizar regressão, são eles:

*    Linear Regression
*    Polynomial Regression
*    Logistic Regression
*    Principal Components Regression (PCR)
